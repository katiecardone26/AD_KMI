# convert to long form

## mean metrics

```python
lr_long = lr.melt(id_vars = 'P-Value Threshold',
                  value_vars = ['Train AUROC', 'Test AUROC', 'Train AUPRC', 'Test AUPRC', 'Train F1', 'Test F1', 'Train Balanced Accuracy', 'Test Balanced Accuracy'],
                  var_name = 'Metric',
                  value_name = 'Score')

lr_long[['Split', 'Metric']] = lr_long['Metric'].str.split(' ', n = 1, expand = True)
lr_long
```

```python
rf_long = rf.melt(id_vars = 'P-Value Threshold',
                  value_vars = ['Train AUROC', 'Test AUROC', 'Train AUPRC', 'Test AUPRC', 'Train F1', 'Test F1', 'Train Balanced Accuracy', 'Test Balanced Accuracy'],
                  var_name = 'Metric',
                  value_name = 'Score')

rf_long[['Split', 'Metric']] = rf_long['Metric'].str.split(' ', n = 1, expand = True)
rf_long
```

```python
xgboost_long = xgboost.melt(id_vars = 'P-Value Threshold',
                            value_vars = ['Train AUROC', 'Test AUROC', 'Train AUPRC', 'Test AUPRC', 'Train F1', 'Test F1', 'Train Balanced Accuracy', 'Test Balanced Accuracy'],
                            var_name = 'Metric',
                            value_name = 'Score')

xgboost_long[['Split', 'Metric']] = xgboost_long['Metric'].str.split(' ', n = 1, expand = True)
xgboost_long
```

## test metrics

```python
xgboost_test_auroc_long = xgboost_test_auroc.drop(columns = ['Unnamed: 0', 'TEST_AUROC_MEAN']).transpose().rename(columns = {0 : 'XGBoost'})
rf_test_auroc_long = rf_test_auroc.drop(columns = ['Unnamed: 0', 'TEST_AUROC_MEAN']).transpose().rename(columns = {0 : 'RF'})
lr_test_auroc_long = lr_test_auroc.drop(columns = ['Unnamed: 0', 'TEST_AUROC_MEAN']).transpose().rename(columns = {0 : 'LR'})

test_auroc_long = pd.concat([xgboost_test_auroc_long, rf_test_auroc_long, lr_test_auroc_long], axis = 1)
test_auroc_long['ITER'] = test_auroc_long.index.str.replace('ITER_', '').astype(int)
test_auroc_long.head()
```

```python
xgboost_test_auprc_long = xgboost_test_auprc.drop(columns = ['Unnamed: 0', 'TEST_AUPRC_MEAN']).transpose().rename(columns = {0 : 'XGBoost'})
rf_test_auprc_long = rf_test_auprc.drop(columns = ['Unnamed: 0', 'TEST_AUPRC_MEAN']).transpose().rename(columns = {0 : 'RF'})
lr_test_auprc_long = lr_test_auprc.drop(columns = ['Unnamed: 0', 'TEST_AUPRC_MEAN']).transpose().rename(columns = {0 : 'LR'})

test_auprc_long = pd.concat([xgboost_test_auprc_long, rf_test_auprc_long, lr_test_auprc_long], axis = 1)
test_auprc_long['ITER'] = test_auprc_long.index.str.replace('ITER_', '').astype(int)
test_auprc_long.head()
```

```python
xgboost_test_f1_long = xgboost_test_f1.drop(columns = ['Unnamed: 0', 'TEST_F1_MEAN']).transpose().rename(columns = {0 : 'XGBoost'})
rf_test_f1_long = rf_test_f1.drop(columns = ['Unnamed: 0', 'TEST_F1_MEAN']).transpose().rename(columns = {0 : 'RF'})
lr_test_f1_long = lr_test_f1.drop(columns = ['Unnamed: 0', 'TEST_F1_MEAN']).transpose().rename(columns = {0 : 'LR'})

test_f1_long = pd.concat([xgboost_test_f1_long, rf_test_f1_long, lr_test_f1_long], axis = 1)
test_f1_long['ITER'] = test_f1_long.index.str.replace('ITER_', '').astype(int)
test_f1_long.head()
```

```python
xgboost_test_ba_long = xgboost_test_ba.drop(columns = ['Unnamed: 0', 'TEST_BALANCED_ACCURACY_MEAN']).transpose().rename(columns = {0 : 'XGBoost'})
rf_test_ba_long = rf_test_ba.drop(columns = ['Unnamed: 0', 'TEST_BALANCED_ACCURACY_MEAN']).transpose().rename(columns = {0 : 'RF'})
lr_test_ba_long = lr_test_ba.drop(columns = ['Unnamed: 0', 'TEST_BALANCED_ACCURACY_MEAN']).transpose().rename(columns = {0 : 'LR'})

test_ba_long = pd.concat([xgboost_test_ba_long, rf_test_ba_long, lr_test_ba_long], axis = 1)
test_ba_long['ITER'] = test_ba_long.index.str.replace('ITER_', '').astype(int)
test_ba_long.head()
```

# combine all test metrics

```python
test_f1_long['Metric'] = 'Test F1 Score'
test_auroc_long['Metric'] = 'Test AUROC'
test_auprc_long['Metric'] = 'Test AUPRC'
test_ba_long['Metric'] = 'Test Balanced Accuracy'

all_test_metrics = pd.concat([test_auroc_long, test_auprc_long, test_ba_long, test_f1_long])
all_test_metrics_long = all_test_metrics.melt(id_vars = ['ITER', 'Metric'], value_vars = ['XGBoost', 'RF', 'LR'], var_name = 'Model', value_name = 'Score')
all_test_metrics_long
```

## test metrics

```python
fig, axes = plt.subplots(2, 2, figsize = (10, 8))
axes = axes.flatten()

for ax, (df, metric) in zip(axes, [(test_auroc_long, 'Test AUROC'), 
                                    (test_auprc_long, 'Test AUPRC'),
                                    (test_ba_long, 'Test Balanced Accuracy'), 
                                    (test_f1_long, 'Test F1 Score')]):
    melted = df.melt(value_vars = ['XGBoost', 'RF', 'LR'], var_name = 'Model', value_name = 'Score')
    sns.boxplot(data = melted, x = 'Model', y = 'Score', hue = 'Model',
                palette = {'XGBoost': '#0072B2', 'RF': '#E69F00', 'LR': '#009E73'}, ax = ax)
    ax.set_title(metric)
    ax.set_xlabel('')

handles = [plt.Rectangle((0 ,0), 1, 1, color = '#0072B2'),
           plt.Rectangle((0, 0), 1, 1, color = '#E69F00'),
           plt.Rectangle((0, 0), 1, 1, color = '#009E73')]
fig.legend(handles, ['XGBoost', 'RF', 'LR'], loc = 'upper center', ncol = 3, bbox_to_anchor = (0.5, -0.02))
fig.suptitle('ADSP Model Test Metrics Across Iterations', fontsize = 16, fontweight = 'bold')

plt.tight_layout()
plt.savefig('AD_KMI/ML/statistical_models/plots/ADSP.test_metrics.across_iterations.boxplot.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

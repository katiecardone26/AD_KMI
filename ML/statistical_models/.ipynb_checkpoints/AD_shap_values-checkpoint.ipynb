# split dataset

### random

```python
avg_pathway_train = avg_pathway.sample(frac = 0.7, random_state = 73)
avg_pathway_test = avg_pathway.drop(avg_pathway_train.index)
```

```python
print(avg_pathway_train.shape)
print(avg_pathway_test.shape)
```

# scale data

```python
predictors = avg_pathway_train.columns.tolist()
predictors.remove('ID')
predictors.remove('AD')

scaler = StandardScaler()
X_avg_pathway_train_scaled = scaler.fit_transform(avg_pathway_train[predictors])
X_avg_pathway_test_scaled = scaler.transform(avg_pathway_test[predictors])
```

# hyperparameter tuning

```python
iter = 73

model = xgb.XGBClassifier(objective = 'binary:logistic', random_state = iter, n_jobs = -1)

n_pos = sum(avg_pathway_train['AD'] == 1)
n_neg = sum(avg_pathway_train['AD'] == 0)
        
param_grid = {
            'n_estimators' : [30, 50, 100, 200, 300, 400, 500, 600],
            'max_depth' : [1, 2, 3],
            'learning_rate' : [0.01, 0.02, 0.03],
            'subsample' : [0.4, 0.5, 0.6, 0.7],
            'colsample_bytree' : [0.4, 0.5, 0.6, 0.7, 0.8],
            'gamma' : [0.3, 0.4, 0.5],
            'min_child_weight' : [20, 30, 40],
            'scale_pos_weight' : [n_neg/n_pos]
}

avg_pathway_random_search = RandomizedSearchCV(estimator = model, param_distributions = param_grid, cv = 5, random_state = iter, scoring = 'balanced_accuracy', n_jobs = -1, n_iter = 10)
avg_pathway_random_search.fit(X_avg_pathway_train_scaled, avg_pathway_train['AD'])

avg_pathway_best_model = avg_pathway_random_search.best_estimator_

y_train_pred_bin = avg_pathway_best_model.predict(X_avg_pathway_train_scaled)
y_train_pred_cont = avg_pathway_best_model.predict_proba(X_avg_pathway_train_scaled)[:, 1]

y_train_pred_bin = avg_pathway_best_model.predict(X_avg_pathway_train_scaled)
y_train_pred_cont = avg_pathway_best_model.predict_proba(X_avg_pathway_train_scaled)[:, 1]

avg_pathway_train_auroc = roc_auc_score(avg_pathway_train['AD'], y_train_pred_cont)
avg_pathway_train_auprc = average_precision_score(avg_pathway_train['AD'], y_train_pred_cont)
avg_pathway_train_f1 = f1_score(avg_pathway_train['AD'], y_train_pred_bin)
avg_pathway_train_balanced_acc = balanced_accuracy_score(avg_pathway_train['AD'], y_train_pred_bin)

print(avg_pathway_train_auroc)
print(avg_pathway_train_auprc)
print(avg_pathway_train_f1)
print(avg_pathway_train_balanced_acc)
```

```python
print(avg_gene_all_train_auroc['ITER_73'])
print(avg_gene_all_train_auprc['ITER_73'])
print(avg_gene_all_train_f1['ITER_73'])
print(avg_gene_all_train_balanced_acc['ITER_73'])
```

# test model

```python
y_test_pred_bin = avg_pathway_best_model.predict(X_avg_pathway_test_scaled)
y_test_pred_cont = avg_pathway_best_model.predict_proba(X_avg_pathway_test_scaled)[:, 1]

avg_pathway_test_auroc = roc_auc_score(avg_pathway_test['AD'], y_test_pred_cont)
avg_pathway_test_auprc = average_precision_score(avg_pathway_test['AD'], y_test_pred_cont)
avg_pathway_test_f1 = f1_score(avg_pathway_test['AD'], y_test_pred_bin)
avg_pathway_test_balanced_acc = balanced_accuracy_score(avg_pathway_test['AD'], y_test_pred_bin)

print(avg_pathway_test_auroc)
print(avg_pathway_test_auprc)
print(avg_pathway_test_f1)
print(avg_pathway_test_balanced_acc)
```

## subset omics file

```python
omics_sub = omics_gene.drop(columns = ['BETA', 'PVAL'])
omics_sub = omics_sub.rename(columns = {'PATHWAY_ID' : 'Feature'})
omics_sub = omics_sub[omics_sub['Feature'].isin(shap_summary['Feature'])]
print(len(shap_summary.index))
print(len(omics_sub.index))
omics_sub.head()
```

## merge

```python
shap_pathway_info = omics_sub.merge(shap_summary, how = 'outer').sort_values("Mean Absolute SHAP Value", ascending = False)
shap_pathway_info
```

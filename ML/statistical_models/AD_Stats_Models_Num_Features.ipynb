# read in input files, clean, and get mean number of features

## rf

```python
rf_avg_gene_standard = pd.read_csv('ML/statistical_models/rf_output/Selected_Features.RF.avg_gene.standard_scaled.go.keep_quest_comb.txt',
                                   sep = '\t')
rf_avg_gene_standard = rf_avg_gene_standard.rename(columns = {'Unnamed: 0' : 'PATHWAY_SCORE'})
rf_avg_gene_standard['PATHWAY_SCORE'] = rf_avg_gene_standard['PATHWAY_SCORE'] + '_standard'
rf_avg_gene_standard = rf_avg_gene_standard.set_index('PATHWAY_SCORE', drop = True)
rf_avg_gene_standard = rf_avg_gene_standard.transpose()
rf_avg_gene_standard['N_FEATURES'] = rf_avg_gene_standard['avg_gene_standard'].str.split(r'\s*,\s*').str.len()
print(rf_avg_gene_standard.shape)
print(rf_avg_gene_standard['N_FEATURES'].mean())
rf_avg_gene_standard.head()
```

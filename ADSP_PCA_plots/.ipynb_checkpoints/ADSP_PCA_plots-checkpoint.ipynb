# load packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# read in input files

```python
gia = pd.read_csv('ADSP_Projects/QC_ADSPv11/PCA/GIA.ancestry',
                  sep = ' ')
print(gia.shape)
gia.head()
```

```python
eigenvec = pd.read_csv('ADSP_Projects/QC_ADSPv11/ADSPphenotype_forAnalysis.txt',
                       sep = '\t')
print(eigenvec.shape)
eigenvec.head()
```

# merge gia and PCA files

```python
eigenvec_gia = eigenvec.rename(columns = {'IID' : 'ID'})
print(len(eigenvec_gia.index))
eigenvec_gia = eigenvec_gia.merge(gia, on = 'ID')
print(len(eigenvec_gia.index))
eigenvec_gia.head()
```

# make plot

```python
color_map = {
    'EUR': 'blue',
    'EAS': 'orange',
    'AFR': 'green',
    'AMR': 'red',
    'MID': 'purple',
    'SAS': 'brown',
    'Other': 'gray'
}

sns.scatterplot(data = eigenvec_gia, x = 'PC1', y = 'PC2', hue = 'race', palette = color_map)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title = "Genetic Similarity Group")
plt.title('ADSP PC1 vs. PC2 Plot')
plt.savefig('AD_KMI/common_var_gene_score/ADSP.PCA.PC1vsPC2_plot.png', bbox_inches = 'tight', dpi = 600)
plt.show()
```

# identify rows that are not equal

```python
new_biospec_sub = new_biospec.drop(columns = ['samplingAgeUnits'])
new_biospec_sub = new_biospec_sub[old_biospec.columns]
```

```python
old_biospec_filt = old_biospec[~old_biospec['individualID'].isin(['GISpool'])]
old_biospec_filt.sort_values(by = ['individualID', 'specimenID'], inplace = True)
old_biospec_filt.head()
```

```python
diff_rows = old_biospec[~old_biospec.eq(new_biospec_sub)].dropna(how='all')
print(len(diff_rows.index))
print(len(old_biospec.index))
diff_rows.head()
```

```python
diff_rows2 =new_biospec_sub[~new_biospec_sub.eq(old_biospec)].dropna(how='all')
print(len(diff_rows2.index))
print(len(new_biospec.index))
diff_rows2.head()
```
